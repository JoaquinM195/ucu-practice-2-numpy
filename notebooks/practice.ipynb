```python
import matplotlib.pyplot as plt
import numpy as np
```

# Distribución Normal

Se hace el randomizador de los números para la distribución normal y posteriormente se hacen los cálculos estadísticos correspondientes

```python
normal = np.random.normal(10,5,1000)

media_normal = np.mean(normal)
mediana_normal = np.median(normal)
varianza_normal = np.var(normal)
desvio_normal = np.std(normal)
minimo_normal = np.min(normal)
maximo_normal = np.max(normal)
```

# Distribución Uniforme
Ídem que en la anterior pero ahora aplicado a la Distribución Uniforme

```python
uniforme = np.random.uniform(0,20,1000)

media_uniforme = np.mean(uniforme)
mediana_uniforme = np.median(uniforme)
varianza_uniforme = np.var(uniforme)
desvio_uniforme = np.std(uniforme)
minimo_uniforme = np.min(uniforme)
maximo_uniforme = np.max(uniforme)
```

# Resultados

Se muestran los resultados de los cálculos estadísticos realizados en los bloques de código anteriores

```python
resultados = {
    "Distribución Normal": {
        "Mediana": mediana_normal,
        "Varianza": varianza_normal,
        "Desvío estándar": desvio_normal,
        "Mínimo": minimo_normal,
        "Máximo": maximo_normal
    },
    "Distribución Uniforme": {
        "Mediana": mediana_uniforme,
        "Varianza": varianza_uniforme,
        "Desvío estándar": desvio_uniforme,
        "Mínimo": minimo_uniforme,
        "Máximo": maximo_uniforme
    }
}

resultados
```

# Gráfico

Se realizan las expresiones gráficas de ambas distribuciones para una mejor apreciación de cada figura

```python
plt.figure(figsize=(13, 8))

#Subplot izquierdo: Normal
plt.subplot(1, 2, 1)
plt.hist(normal, bins='sqrt')
plt.title("Histograma - Distribución Normal")
plt.xlabel("Contenedores, agrupaciones")
plt.ylabel("Frecuencia, cantidad")
plt.grid(axis='y')


#Subplot derecho: Uniforme
plt.subplot(1, 2, 2)
plt.hist(uniforme, bins='sqrt',color='red')
plt.title("Histograma - Distribución Uniforme")
plt.xlabel("Contenedores, agrupaciones")
plt.ylabel("Frecuencia, cantidad")
plt.grid(axis='y')

plt.suptitle("Comparación de Distribuciones", fontsize=14)
plt.tight_layout()
plt.show()
```
